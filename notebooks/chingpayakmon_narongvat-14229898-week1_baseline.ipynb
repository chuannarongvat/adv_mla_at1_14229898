### Categorical Columns

We will extract the categorical columns from our df.

```python
cat_cols = [feature for feature in df_cleaned.columns if df_cleaned[feature].dtypes=='O']
```

```python
num_cols = [feature for feature in df_cleaned.columns if df_cleaned[feature].dtypes!='O']
```

```python
cat_cols
```

The following columns fall under the categorical category:

- **team:** Name of team
- **conf:** Name of conference
- **yr:** Students year of study: `Fr` for freshmen, `So` for sophomores, `Jr` for juniors, `Sr` for seniors
- **ht:** Height of student
- **num:** Player's number
- **type:** Type of metrics displayed: `All` for all types, `C` for conference`, `NC` for non-conference, `PC` for pre-conference tour, `R` for regular season, `P` for post-season, `T` for NCAA
- **player_id:** Unique identifier of player.


```python
for i in cat_cols:
    print(f"{i}: {train_df[i].nunique()}")
```

We will perform the following steps for categorical columns:

1. **One-Hot Encoding:** We'll perform one hot encoding on 'conf' columns.
2. **Hash-Encoding:** We'll perform hash encoding on 'team' columns.
3. **Ordianal Mapping:** We'll convert the ordinal category into numerical values, mapping `Fr` to 1, `So` to 2, `Jr` to 3, and `Sr` to 4.
4. **Drop some columns:** 'ht' column will be drop, since the height column contain mixed data type and didn't provide any useful information. 'type' and 'player_id' will also be dropped, since 'type' column has only one unique value and 'player_id' column is a unique identifier for each player.

5. **Other Technique:**

#### One-hot encoding

#### Ordinal Mapping

```python
df_cleaned['yr'].isna().sum()
```

```python
df_cleaned['yr'].value_counts()
```

The 'yr' columns represent the student year's of study:
- Jr: Junoirs
- Fr: Freshmen
- So: Sophomores
- Sr: Seniors

There are some mixture of numerics data types, and in this case, we will filter out the rows that don't belong in these categories.

```python
valid_yr_cat = ['Jr', 'Fr', 'So', 'Sr']
df_cleaned = df_cleaned[df_cleaned['yr'].isin(valid_yr_cat)]
```

```python
yr_mapping = {
    'Fr': 1,
    'So': 2,
    'Jr': 3,
    'Sr': 4
}
```

```python
df_cleaned['yr'] = df_cleaned['yr'].map(yr_mapping)
```

```python
df_cleaned['yr'].value_counts()
```

#### Drop 'type', 'ht', and 'player_id'

```python
df_cleaned['type'].value_counts()
```

```python
df_cleaned['ht'].value_counts()
```

```python
df_cleaned.drop(columns=['type', 'player_id', 'ht'], inplace=True)
```

#### Label Encoding on 'num'

```python
df_cleaned['num'].isna().sum()
```

```python
df_cleaned['num'].dtype
```

```python
df_cleaned['num'].value_counts()
```

```python
special_case = {'--': '0', '31/24': '31', '23B': '23', '4A': '4'}
df_cleaned['num'] = df_cleaned['num'].replace(special_case)
```

```python
df_cleaned['num'] = df_cleaned['num'].str.split('.').str[0]
```

```python
df_cleaned['num'].value_counts()
```

```python
df_cleaned.shape    
```

```python
df_cleaned['num'].fillna('Missing', inplace=True)
```

```python
encoded_num = pd.get_dummies(df_cleaned['num'], prefix='num')
encoded_num = encoded_num.astype(int)
```

```python
df_cleaned = pd.concat([df_cleaned, encoded_num], axis=1)
df_cleaned.drop(columns=['num'], inplace=True)
```

```python
fig, axes = plt.subplots(nrows=len(missing_pattern), figsize=(10, 20))

for i,c in enumerate(missing_pattern):
    sns.histplot(df_cleaned[c], ax=axes[i], kde=True, color='orange')
    
    mean = df_cleaned[c].mean()
    median = df_cleaned[c].median()
    
    axes[i].axvline(mean, color='red', linestyle='--', label=f"Mean: {mean:.2f}")
    axes[i].axvline(median, color='blue', linestyle='--', label=f"Median: {median:.2f}")
    
    axes[i].set_title(f'Histrogram for {c}')
    axes[i].set_xlabel(c)
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
